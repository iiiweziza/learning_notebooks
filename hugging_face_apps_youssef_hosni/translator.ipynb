```python
from transformers import pipeline
```

```python
translator = pipeline("translation", model="facebook/nllb-200-distilled-600M")
```

```python
text = "Learning AI involves studying machine learning, deep learning, and data science. It also helps to practice coding with libraries such as TensorFlow and PyTorch."
print(text)
```

```python
response = translator(text,max_length=400,
                      src_lang='eng_Latn',
                      tgt_lang='arb_Arab')
print(response)
```
